```python
import torch
from torch import nn, optim
from jcopdl.callback import Callback, set_config
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
device= torch.device("cuda"if torch.cuda.is_available() else "cpu")
device
```

# Import Data

```python
df=pd.read_csv("data/Sunspots.csv",index_col="Date",parse_dates=["Date"])
df
```

# Dataset Splitting

```python
from sklearn.model_selection import train_test_split
```

```python

ts_train, ts_test = train_test_split(df, test_size=0.2,shuffle=False)
ts_train.shape, ts_test.shape
```

# Dataset and Dataloader

```python
from jcopdl.utils.dataloader import TimeSeriesDataset
from torch.utils.data import DataLoader
```

```python
seq_len= 12
bs=64

train_set= TimeSeriesDataset(ts_train,"Monthly_Mean_Sunspot",seq_len)
trainloader=DataLoader(train_set,batch_size=bs)

test_set=TimeSeriesDataset(ts_test,"Monthly_Mean_Sunspot",seq_len)
testloader= DataLoader(test_set,batch_size=bs)
```

# Arsitektur and Config

```python
class LSTM (nn.Module):
    def __init__(self,input_size,output_size, hidden_size, num_layers, dropout):
        super().__init__()
        self.hidden= nn.LSTM(input_size, hidden_size, num_layers, dropout=dropout, batch_first=True)
        self.fc = nn.Linear(hidden_size, output_size)
        
        
    def forward(self,x,hidden):
        x,hidden= self.hidden(x,hidden)
        x= self.fc(x)
        return x,hidden
```

```python
config = set_config({
    "batch_size" : bs,
    "seq_len" : train_set.n_seq,
    "input_size" : train_set.n_feature,
    "output_size" : 1,
    "hidden_size" : 64,
    "num_layers" : 2,
    "dropout": 0
})
```

# Training Preparation ~> MCOC

```python
model = LSTM(config.input_size,config.output_size,config.hidden_size,config.num_layers,config.dropout).to(device)
criterion = nn.MSELoss(reduction="mean")
optimizer = optim.AdamW(model.parameters(), lr=0.001)
callback = Callback(model,config,outdir="model")
```

# Training Loop

```python
def loop_fn (mode, dataset, dataloader, model, criterion, optimizer, device):
    if mode =="train":
        model.train()
    elif mode == "test":
        model.eval()
        
    cost = 0
    for feature,target in tqdm(dataloader,desc = mode.title()):
        feature, target = feature.to(device), target.to(device)
        output, hidden = model(feature,None) 
        loss = criterion(output, target)
        
        if mode == "train":
            loss.backward()
            optimizer.step()
            optimizer.zero_grad()
            
        cost += loss.item()*feature.shape[0]
    cost=cost/len(dataset)
    return cost
```

```python
from tqdm.auto import tqdm
```

```python
while True :
    train_cost= loop_fn("train",train_set,trainloader,model,criterion,optimizer,device)
    with torch.no_grad():
        test_cost= loop_fn("test",test_set,testloader,model,criterion,optimizer,device)

    callback.log(train_cost, test_cost)


    callback.save_checkpoint()


    callback.cost_runtime_plotting()



    if callback.early_stopping(model, monitor='test_cost'):
        callback.plot_cost()
        break
```
